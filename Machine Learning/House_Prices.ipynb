```python
order_MSZoning = ["RH", "RM", "RL", "RP", "I", "FV", "A", "C (all)"]
training_data["MSZoning"] = training_data["MSZoning"].apply(lambda x : order_MSZoning.index(x))

order_LotShape = ["IR3", "IR2", "IR1", "Reg"]
training_data["LotShape"] = training_data["LotShape"].apply(lambda x : order_LotShape.index(x))

order_Utilities = ["ELO", "NoSeWa", "NoSewr", "AllPub"]
training_data["Utilities"] = training_data["Utilities"].apply(lambda x : order_Utilities.index(x))

LandSlope = ["Sev", "Mod", "Gtl"]
training_data["LandSlope"] = training_data["LandSlope"].apply(lambda x : LandSlope.index(x))

BldgType = ["1Fam", "2fmCon", "Duplex", "TwnhsE", "Twnhs"]
training_data["BldgType"] = training_data["BldgType"].apply(lambda x : BldgType.index(x))


HouseStyle = ["SLvl", "SFoyer", "1.5Fin", "1Story" ,"1.5Unf", "2.5Unf", "2.5Fin", "2Story" ]
training_data["HouseStyle"] = training_data["HouseStyle"].apply(lambda x : HouseStyle.index(x))

order_ExterQual = ["Po", "Fa", "TA", "Gd", "Ex"]
training_data["ExterQual"] = training_data["ExterQual"].apply(lambda x : order_ExterQual.index(x))

order_ExterCond = ["Po", "Fa", "TA", "Gd", "Ex"]
training_data["ExterCond"] = training_data["ExterCond"].apply(lambda x : order_ExterCond.index(x))

order_Foundation = ["Wood", "Stone", "Slab", "PConc", "CBlock", "BrkTil"]
training_data["Foundation"] = training_data["Foundation"].apply(lambda x : order_Foundation.index(x))

order_BsmtQual = ["NA","Po", "Fa", "TA", "Gd", "Ex"]
training_data["BsmtQual"] = training_data["BsmtQual"].apply(lambda x : order_BsmtQual.index(x))

order_BsmtCond = ["NA","Po", "Fa", "TA", "Gd", "Ex"]
training_data["BsmtCond"] = training_data["BsmtCond"].apply(lambda x : order_BsmtCond.index(x))

CentralAir = ["N", "Y"]
training_data["CentralAir"] = training_data["CentralAir"].apply(lambda x : CentralAir.index(x))

Electrical  = ["Mix", "FuseP", "FuseF", "FuseA", "SBrkr"]
training_data["Electrical"] = training_data["Electrical"].apply(lambda x : Electrical.index(x))

KitchenQual = ["Po", "Fa", "TA", "Gd", "Ex"]
training_data["KitchenQual"] = training_data["KitchenQual"].apply(lambda x : KitchenQual.index(x))

GarageQual = ["NA","Po", "Fa", "TA", "Gd", "Ex"]
training_data["GarageQual"] = training_data["GarageQual"].apply(lambda x : GarageQual.index(x))

GarageCond = ["NA","Po", "Fa", "TA", "Gd", "Ex"]
training_data["GarageCond"] = training_data["GarageCond"].apply(lambda x : GarageCond.index(x))

SaleType = ["Oth", "ConLD", "ConLI", "ConLw", "Con", "COD", "New", "VWD", "CWD", "WD"]
training_data["SaleType"] = training_data["SaleType"].apply(lambda x : SaleType.index(x))

SaleCondition = ["Partial", "Family", "Alloca", "AdjLand", "Abnorml", "Normal"]
training_data["SaleCondition"] = training_data["SaleCondition"].apply(lambda x : SaleCondition.index(x))
```

```python
non_numerical_columns = training_data.select_dtypes(exclude = np.number).columns
non_numerical_columns
```

### Splitting Data

```python
X = training_data.drop(["SalePrice"], axis = 1)
Y = training_data["SalePrice"]
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression

lg = LinearRegression()

lg.fit(X_train, Y_train)

print(str(round(lg.score(X_test, Y_test) * 10000) / 100) + "% : Confidence")
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(X_train, Y_train)

print(str(round(forest.score(X_test, Y_test) * 10000) / 100) + "% : Confidence")
```
